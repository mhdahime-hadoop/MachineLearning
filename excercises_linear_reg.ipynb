```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv(r'D:\MitStudy\Python\FullStackDataScience\DataProjects\Pandas\datasets\excercise\linear_reg\canada_per_capita_income.csv')
df.tail(3)
```

```python
plt.scatter(df.year,[df['per capita income (US$)']], marker='*')
plt.xlabel('year', color='g', size=15)
plt.ylabel('per capita income (US$)', color='g', size=15)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
plt.scatter(df.year,[df['per capita income (US$)']], marker='*')
plt.xlabel('Year', color='g', size=15)
plt.ylabel('Income', color='g', size=15)
plt.plot(df.year, reg.predict(df[['year']]), color='g')
```
