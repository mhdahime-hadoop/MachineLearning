```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\MitStudy\Machine_Learning_Tutorial_and_exercise\Prectice\car_evaluation\car.data')
df.head()
```

```python
df.columns = ['buying', 'maint', 'door', 'persons', 'lug_boot', 'safety', 'target']
```

```python
df.tail(30)
```

```python
features = df.drop('target',axis = 'columns')
```

```python
labels = df['target']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_buying = LabelEncoder()
le_maint = LabelEncoder()
le_door = LabelEncoder()
le_persons = LabelEncoder()
le_lug_boot = LabelEncoder()
le_safety = LabelEncoder()
```

```python
features['buying_v'] = le_buying.fit_transform(features['buying'])
features['maint_v'] = le_buying.fit_transform(features['maint'])
features['door_v'] = le_buying.fit_transform(features['door'])
features['persons_v'] = le_buying.fit_transform(features['persons'])
features['lug_boot_v'] = le_buying.fit_transform(features['lug_boot'])
features['safety_v'] = le_buying.fit_transform(features['safety'])
```

```python
features.head()
```

```python
inputs = features.drop(['buying','maint','door', 'persons', 'lug_boot', 'safety'], axis = 'columns')
inputs.tail(3)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,labels)
```

```python
model.score(inputs,labels)
```

```python
model.predict([[1,1,3,2,0,0]])
```
